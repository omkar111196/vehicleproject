```python
#Number of cars available by cities
cars=pd.read_csv("Dataset - Used cars.csv")
cities=cars['Location'].value_counts()
print(cities)
```

```python
plt.plot(cities, linestyle="dotted", color='r', marker='o')
plt.title('City wise available cars')
plt.show()
```

<h3><b>So we came to know that MUMBAI has the most demand of cars

```python
#Number of cars available by year
cars=pd.read_csv("Dataset - Used cars.csv")
available = cars['Year'].value_counts()
print(available)
```

```python
import plotly.express as px
fig = px.pie(available, names=cars['Year'], title='Year wise distribution of Cars manufactured',)
fig.show()
```

<h3><b>Here we can say that the most sales of cars were made in <u>2014

```python
#Number of cars available by fuel
cars=pd.read_csv("Dataset - Used cars.csv")
fuel = cars['Fuel_Type'].value_counts()
print(fuel)
```

```python
fig = px.scatter(cars, x="Fuel_Type", y="Price", size="Year", color=cars['Price'],size_max=50,
                title=' ')
fig.show()
```

<h3><b>'Most' People prefer Diesel Cars followed by Petrol Cars. CNG and LPG Cars are 'least' prefered
